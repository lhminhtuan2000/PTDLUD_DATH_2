# EDA

```python
# Đọc dữ liệu nguồn 1
df_1 = pd.read_csv(#"/Users/admin/Downloads/Listings.csv"
    "https://raw.githubusercontent.com/adishourya/Airbnb/master/new-york-city-airbnb-open-data/AB_NYC_2019.csv"
)
df_1.head()
```

```python
# Đọc dữ liệu nguồn 2
df_2 = pd.read_csv("/Users/admin/Downloads/airbnb-listings.csv", sep=';', low_memory=False)
df_2
```

```python
# Lấy dòng dữ liệu liên quan đến New York của df_2
df_2 = df_2[df_2['City'] == 'New York']
df_2
```

## Gộp 2 nguồn dữ liệu

```python
# format thống nhất tên cột 2 nguồn dữ liệu
df_1.columns = (df_1.columns.str.upper())
df_2.columns = (df_2.columns.str.replace(' ', '_').str.upper())
```

```python
# Lấy những cột chung ở 2 nguồn dữ liệu
selected_cols = df_1.columns.intersection(df_2.columns)
selected_cols
# print(df_1.columns.difference(df_2.columns))
```

```python
# Xoá các cột không cần thiết
selected_cols = selected_cols.drop(['NAME', 'HOST_NAME'])
selected_cols
```

```python
# Gộp 2 data frame
df = pd.concat([df_1[selected_cols], df_2[selected_cols]])
# Reset index
df.reset_index(inplace=True)
# Xoá index cũ
df.drop(columns=['index'], inplace=True)
df
```

```python
# Kiểm tra dữ liệu trùng
df.duplicated().sum()
```
